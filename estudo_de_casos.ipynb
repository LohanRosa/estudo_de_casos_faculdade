# Big Data Aplicado à Indústria Automobilística: Vendas e Comportamento de Compra

Com o crescimento exponencial da geração de dados, a indústria automobilística tem se beneficiado amplamente da aplicação de Big Data para compreender melhor o mercado, otimizar estratégias de vendas e identificar padrões de comportamento de compra. Neste estudo de caso, utilizamos técnicas de análise de dados para explorar um grande volume de informações relacionadas a vendas de veículos, com o objetivo de obter insights relevantes que possam orientar decisões de negócio.

Nosso foco está voltado para quatro frentes principais:

* Limpeza e manutenção dos dados: etapa essencial para garantir a confiabilidade das análises, onde tratamos valores ausentes, duplicados e inconsistências no conjunto de dados.
* Análise estatística: aplicação de métricas descritivas para identificar tendências, variações e correlações relevantes entre variáveis como tipo de veículo, faixa de preço, entre outros.
* Visualização de dados com Seaborn e Matplotlib: criação de gráficos e diagramas que facilitam a interpretação dos dados e destacam padrões e anomalias de forma intuitiva e informativa.
* Respostas a perguntas de negócios: formulação e resposta a questões estratégicas

Ao longo do estudo, combinamos métodos estatísticos e ferramentas de visualização para construir uma narrativa baseada em dados, com foco em na venda de carros no Brasil nos anos de 2022, 2023, 2024.

# Visualização 

Agora que verificamos alguns dados bases Estatísticos, vamos fazer algumas visualizações
 

```python
vendas_por_ano = todos_os_dados[todos_os_dados['Ano'].isin([2022, 2023, 2024])].groupby('Ano')['Unidades Vendidas'].sum()
```

```python
# grafica as vendas por ano 
plt.figure(figsize=(8, 5))
plt.plot(vendas_por_ano.index, vendas_por_ano.values, marker='o', linestyle='-', color='b')
plt.title('Unidades Vendidas por Ano')
plt.xlabel('Ano')   
plt.ylabel('Unidades Vendidas')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# df com colunas: 'ano', 'preco'
preco_medio_por_ano = todos_os_dados[todos_os_dados['Ano'].isin([2022, 2023, 2024])].groupby('Unidades Vendidas')['Preço Médio Estimado (R$)'].sum()
```

```python
# plotando o preço médio por ano
plt.figure(figsize=(10 , 6))
plt.plot(preco_medio_por_ano.index, preco_medio_por_ano.values, marker='o', linestyle='-', color='g')
plt.title('Preço Médio por unidade Vendida')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Preço Médio Estimado (R$)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Nesse gráfico conseguimos analisar que os a maior quantidade de dados estão até quantile de valor 40 mil unidades vendidas, mostrando assim que temos sim muitas unidades
que se encontram após porque são os top 10 carros vendidos a cada ano como Strada, HB20, Polo e etc, fazendo com que essas marcas se tornem outlier situando muito longe dos outros pontos de dados de um conjunto. vamos ver a quantidade vendida por ano dos top 3 veiculos 

```python
# plotando as vendas totais por ano (teste de visualização)
strada = todos_os_dados[todos_os_dados['Modelo'] == 'Fiat Strada']
plt.figure(figsize=(8,5))
sns.barplot(x='Ano', y='Unidades Vendidas', data=strada)
plt.title('Vendas da Fiat Strada (2022–2024)')
plt.ylabel('Unidades Vendidas')
plt.show()
```

O modelo Fiat Strada é primeiro lugar de todos os anos, já respondendo uma das perguntas de negócio: Qual o carro mais vendido dos 3 anos? e qual categoria?
Strada, pois a categoria dele ser picape nos informa que muitas empresas privadas ou particulares compram para colocar em frotas de trabalho. 

```python
hb20 = todos_os_dados[todos_os_dados['Modelo'] == 'Hyundai HB20']
plt.figure(figsize=(8,5))
sns.barplot(x='Ano', y='Unidades Vendidas', data=hb20)
plt.title('Vendas do Hyundai HB20 (2022–2024)')
plt.ylabel('Unidades Vendidas')
plt.show()
```

HB20 sendo o segundo por ser hatch e seu preço "acessivel" aos demais e com uma confiabilidade que conquistou aos longo dos anos, possivelmente carros para familias pequenas e pessoas solteiras ou estudantes de universidade. 

```python
onix = todos_os_dados[todos_os_dados['Modelo'] == 'Chevrolet Onix']
plt.figure(figsize=(8,5))
sns.barplot(x='Ano', y='Unidades Vendidas', data=onix)
plt.title('Vendas do Chevrolet Onix (2022–2024)')
plt.ylabel('Unidades Vendidas')
plt.show()
```

Onix logo seguido por Onix Plus se encontra em terceiro lugar pelo preço e pelo plus ser um sedan assim se tornando um carro mais para familia.
